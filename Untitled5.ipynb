```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hotels.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Rating'],df['Price'])
plt.xlabel('Rating')
plt.ylabel('Price')
```

```python
# Filter out rows where 'Price' or 'Rating' are NaN to ensure numerical data for regression
df_cleaned = df.dropna(subset=['Rating', 'Price'])

# Select 'Rating' as the feature (X) and 'Price' as the target (Y)
x = df_cleaned[['Rating']]
y = df_cleaned['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
br= LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
br.fit(X_train,Y_train)
```

```python
X_test
```

```python
br.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['Rating'],df['Price'])
plt.plot(X_train,br.predict(X_train),color='red')
plt.xlabel('Price(in dollar)')
plt.ylabel('Rating')
```

```python
m = br.coef_
m
```

```python
b = br.intercept_
b
```
